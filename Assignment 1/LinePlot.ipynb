```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
rep0 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep0\fitstats.csv', delimiter = ',', index_col = 0)
rep1 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep1\fitstats.csv', delimiter = ',', index_col = 0)
rep2 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep2\fitstats.csv', delimiter = ',', index_col = 0)
rep3 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep3\fitstats.csv', delimiter = ',', index_col = 0)
rep4 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep4\fitstats.csv', delimiter = ',', index_col = 0)
rep5 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep5\fitstats.csv', delimiter = ',', index_col = 0)
rep6 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep6\fitstats.csv', delimiter = ',', index_col = 0)
rep7 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep7\fitstats.csv', delimiter = ',', index_col = 0)
rep8 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep8\fitstats.csv', delimiter = ',', index_col = 0)
rep9 = pd.read_csv(r'C:\Evoluation_computing\multi_enemy3\rep9\fitstats.csv', delimiter = ',', index_col = 0)
```

```python
rep = [rep0,rep1,rep2,rep3,rep4,rep5,rep6,rep7,rep8,rep9]
```

```python
def std_list(rep,column):
    mylist = list()
    for df in rep:
        x = df[column].values
        mylist.append(x)
    return mylist      
```

```python
avg_data = std_list(rep, 'Avg fitness')
best_data = std_list(rep, 'Best fitness')
```

```python
def transform(data):
    all_data = []
    for i in range(21):
        temp = list()
        for arr in data:
            temp.append(arr[i])
        all_data.append(temp)
    return all_data
```

```python
avg_data_transform = transform(avg_data)
best_data_transform = transform(best_data)
```

```python
def cal_std(data):
    results = []
    for arr in data:
        x = np.std(arr)
        results.append(x)
    return results
```

```python
enemy1_avg_std = cal_std(avg_data_transform)
enemy1_best_std = cal_std(best_data_transform)
```

```python
enemy1 = rep0.add(rep1).add(rep2).add(rep3).add(rep4).add(rep5).add(rep6).add(rep7).add(rep8).add(rep9)
enemy1['Best fitness'] = enemy1['Best fitness'].apply(lambda x: x / 10)
enemy1['Avg fitness'] = enemy1['Avg fitness'].apply(lambda x: x / 10)
enemy1[['Best fitness', 'Avg fitness']] = enemy1[['Best fitness', 'Avg fitness']].applymap(lambda x: '%.2f'%x)
enemy1['generation'] = np.arange(0,21,1)
enemy1['std_avg'] = enemy1_avg_std
enemy1['std_best'] = enemy1_best_std
enemy1.drop(columns = ['Std fitness'], inplace = True)
enemy1['Best fitness'] = pd.to_numeric(enemy1['Best fitness'])
enemy1['Avg fitness'] = pd.to_numeric(enemy1['Avg fitness'])
enemy1.to_csv(r'C:\Evoluation_computing\multi_enemy3\multi_enemy3.csv')
#enemy1.head(21)
```

```python
mult_enemy1 = pd.read_csv('Data\mult_enemy1.csv')
double_enemy1 = pd.read_csv(r'C:\Evoluation_computing\double_enemy1\double_enemy1.csv')
```

```python
enemy1 = plt.figure(1)
sns.set_style("darkgrid")

ax1 = sns.lineplot(x='generation', y="Best fitness",err_style = 'band',color = 'blue',data=mult_enemy1,label = '4 parents (max fitness)')
ax1.lines[0].set_linestyle("--")
plt.fill_between(x = 'generation',y1 = mult_enemy1['Best fitness'] - mult_enemy1['std_best'], y2 = mult_enemy1['Best fitness'] + mult_enemy1['std_best'],color='blue', alpha=0.1,data = mult_enemy1)
ax2 = sns.lineplot(x='generation', y="Avg fitness",color = 'red',data=mult_enemy1,label = '4 parents (mean fitness)')
ax2.lines[1].set_linestyle('--')
plt.fill_between(x = 'generation',y1 = mult_enemy1['Avg fitness'] - mult_enemy1['std_avg'], y2 = mult_enemy1['Avg fitness'] + mult_enemy1['std_avg'],color='red', alpha=0.1,data = mult_enemy1)

sns.lineplot(x='generation', y="Best fitness",err_style = 'band',color = 'blue',data=double_enemy1,label = '2 parents (max fitness)', marker = '')
plt.fill_between(x = 'generation',y1 = double_enemy1['Best fitness'] - double_enemy1['std_best'], y2 = double_enemy1['Best fitness'] + double_enemy1['std_best'],color='blue', alpha=0.1,data = double_enemy1)
sns.lineplot(x='generation', y="Avg fitness",color = 'red',data=double_enemy1,label = '2 parents (mean fitness)')
plt.fill_between(x = 'generation',y1 = double_enemy1['Avg fitness'] - double_enemy1['std_avg'], y2 = double_enemy1['Avg fitness'] + double_enemy1['std_avg'],color='red', alpha=0.1,data = double_enemy1)

plt.xlabel('Generation')
plt.ylabel('Fitness')
my_x_ticks = np.arange(0, 21, 1)
plt.xticks(my_x_ticks)
plt.legend(loc='lower right', fontsize = 8)
plt.savefig(r'C:\Evoluation_computing\mult_enemy1\generation_fitness_enemy1_test.png')
```
